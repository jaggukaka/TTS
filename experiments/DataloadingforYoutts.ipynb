```python
from genericpath import exists
os.chdir('/content')

import random, pathlib, csv, shutil

total_samples = 700
num_unique_speakers = 5
min_samples_per_speaker = total_samples/num_unique_speakers

unique_speaker_list = {}
update_only_meta_file = True

destination_dir = '/content/drive/MyDrive/mailabs_new/fr_FR/'

for gender in ['male', 'female'] :
  curdir = os.path.join('/content/fr_FR', gender)
  print (curdir, os.path.isdir(curdir))
  if os.path.isdir(curdir):
    root, dirs, _ = next(os.walk(curdir))
    for dr in dirs:
      speaker = (pathlib.Path(dr).stem)
      if os.path.isdir(os.path.join(root, dr)):
        l1root, l1dirs, _ = next(os.walk(os.path.join(root, dr)))
        totalbooks = len(l1dirs)
        for booktitle in l1dirs:
          sourcepath = os.path.join(l1root, booktitle)
          with open(os.path.join(sourcepath, 'metadata.csv'), "r", encoding="utf-8") as ttf:
            alllines = ttf.readlines()
            selectedlines = random.sample(alllines, int(min_samples_per_speaker/totalbooks))
            predest = os.path.join(destination_dir, gender, speaker, booktitle)
            destpath = os.path.join(predest, 'wavs')
            os.makedirs(destpath, exist_ok=True)
            with open(os.path.join(predest, 'metadata.csv'), 'w+') as destfileobj:
              destfileobj.seek(0)
              for line in selectedlines:
                splits = line.split('|')
                fname = splits[0] + '.wav'
                #print(f"file to write {os.path.join(destpath, splits[0] + '.wav')}")
                try :
                  sourcefilename = os.path.join(sourcepath, 'wavs', fname)
                  shutil.copy(sourcefilename ,os.path.join(destpath, fname) )
                  destfileobj.write(splits[0] + '|' + splits[1] + '\n')
                  if speaker in unique_speaker_list:
                    unique_speaker_list[speaker] += 1
                  else :
                    unique_speaker_list[speaker] = 1
                except :
                 print (f"Skipping file {sourcefilename}")
              destfileobj.close()
      if len(unique_speaker_list) >= min_samples_per_speaker:
        break
              

from functools import reduce
print (reduce(lambda x, y : x + y, unique_speaker_list.values()))
```

```python

def librispeech(root_path, meta_files=None, ignored_speakers=None):
    items = []
    meta_files = glob(f"{root_path}/**/*.trans.txt", recursive=True)
    splitlist = []
    for metafile in meta_files:
      parents = pathlib.Path(metafile).parents
      with open(metafile, 'r+') as mfc:
        for line in mfc:
          metadata = line.split(' ')
          soundfile = os.path.join(parents[0], metadata[0])
          #flacpath = soundfile + '.flac'
          wavpath = soundfile + '.wav'
          #flac_version = AudioSegment.from_file(flacpath, "flac")
          #flac_version.export( wavpath , format="wav")
          text = ' '.join(metadata[1:]).strip()
          speaker = parents[1].stem
          splitlist.append([text[:1] + text[1:].lower(), wavpath, 'librispeech_en_' + speaker])
    

    print('\n'.join([','.join(item) for item in splitlist[:5]]))
    return splitlist

k = librispeech('/content/drive/MyDrive/librispeechmin/LibriSpeech/dev-clean-2')
```
